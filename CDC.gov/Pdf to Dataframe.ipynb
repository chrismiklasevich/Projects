```python
import pandas as pandas
from tabula import read_pdf
import os

#Important function for changing the string values to become floats
def fixnum(s):
    try:
        return float(s)
    except ValueError:
        return str(s)

```

```python
listpdf1 = read_pdf("C:/Users/chrism/Desktop/Projects/CDC.gov/diabetes_card.pdf",pages="11",stream='TRUE')
listpdf2 = read_pdf("C:/Users/chrism/Desktop/Projects/CDC.gov/diabetes_card.pdf",pages="12",stream='TRUE')
```

```python
#listpdfs are within a list. Need to exstract the dataframe
pdf1 = listpdf1[0]
pdf2 = listpdf2[0]
```

```python
#How it looks
pdf1.head()
```

```python
pdf1.rename(columns = {'Table 2. Percentage of US Adults Aged 18 or Older With Diagnosed Diabetes Who Reported':'place'}, inplace=True)
#Renaming the column of the dataframe to be easier to use
```

```python
table1 = []
#THis is page 11 table
for row in pdf1.place[2:]:
        #df[['First','Last']] = df.Name.str.split(expand=True) 

    templist = row.split()
    #print(templist,"LIst")
    #for i in templist:
        #if templist[i].isdigit() == False and templist[i-1].isdigit():
            #templist[i-1:i] = [''.join(templist[i-1:i])]
    stringlist = []
    numberlist = []
    
    for strings in templist:

        if strings.replace('.', '', 1).isdigit() == False:
            stringlist.append(strings)
        else:
            numberlist.append(strings)
        #print(stringlist,"STRING")
        #print(numberlist,"NUMBER")
    #Now need to join the string list together and then combine the string and number list for one list
    #Then start appending correctly
            
    table1.extend(stringlist)
    table1.extend(numberlist)


print(table1[0:30])
```

```python
#Need to change the numbers from strings to floats
tab1fix = []
for i in table1:
    tab1fix.append(fixnum(i))
    #print(type(i))
print(tab1fix[0:15])
```

```python
#For checking indexes that are strings and locating the constant NA values caused by Idaho
nalist = []
for i in range(len(tab1fix)):
    if isinstance(tab1fix[i], str):
        nalist.append(i)
print(nalist)
#Deletes Idaho's NAs
del tab1fix[74:80]

#print(fix)
```

```python
#Combines states/district that are more than 1 word. Example: 'District', 'of', 'Columbia' *or* 'New', 'Hampshire'
try:
    for i in range(len(tab1fix)):
        if isinstance(tab1fix[i],str):
            if isinstance(tab1fix[i+1],str):
                if isinstance(tab1fix[i+2],str):
                    #print("3",fix[i :i+3])
                    tab1fix[i : i+3] = [' '.join(tab1fix[i : i+3])]
                else:
                    #print("ESLSE",fix[i :i+2])
                    tab1fix[i : i+2] = [' '.join(tab1fix[i : i+2])] 
except IndexError:
    print("")
print(tab1fix[220:240])
```

```python
#Setting up the lists to be reformed back into a dataframe of 6 columns
tab1_list = []
v = 0
while v < (len(tab1fix)):
    tab1_list.append([tab1fix[v],tab1fix[v+1],tab1fix[v+2],tab1fix[v+3],tab1fix[v+4],tab1fix[v+5]])
    v = v + 6
tab1_list[0:10]
```

```python
#'Daily Self-by Health State A1c Tests in the Foot Exam Two or more  Monitoring of Professional in  Last Year Blood Glucosethe Last YearManagement Class Ever Attended Diabetes Self-  Exam in the  Dilated Eye Last Year'
#Renameing columns to proper abrivation
df1 = (pandas.DataFrame(tab1_list,columns=['State','2+ A1cTests','Foot Exm','Self-manage class','Dilate Eye','Bld glucose monitor']))
df1.head()
```

```python
#Now looking at pdf1 or page 2 of our table, there numerous columns filled with NAs.
#Remove the columns by name
pdf2.rename(columns = {'Table 2, continued':'place'}, inplace=True)
pdf2rem = pdf2[1:8].drop('Unnamed: 2',1)
pdf2rem = pdf2rem[0:8].drop('Unnamed: 3',1)
pdf2rem = pdf2rem[0:8].drop('Unnamed: 5',1)
pdf2rem = pdf2rem[0:8].drop('Unnamed: 6',1)
pdf2rem.head()

```

```python
#Fixing the indexes/rows of the page of table to start from 0 
#*Important for merging by columns* If not matching indexes for the rows, the values would be misaligned
pdf2_ind = pdf2rem.rename(index={1: 0,2: 1,3: 2,4: 3,5: 4,6: 5,7: 6})
pdf2_ind.head()
```

```python
#Transforming the dataframe column into list for splitting the first column into States and float
str_fix2 = pdf2_ind.place.str.split(expand=True)
listsplit = str_fix2.values.tolist()
listsplit
```

```python
remove_NA = []
try:
    for i in listsplit:
        templist = []
        for j in i:
            #print(templist,'TEMPLIST')
            if j != None:
                templist.append(j)
        remove_NA.append(templist)
except IndexError:
    print("")
print(remove_NA)
#remove_NA is removing the 'None' values. 
```

```python
#Joining 'West' and 'Virgina' together
remove_NA[4][0:2]=[' '.join(remove_NA[4][0 :2])] 
remove_NA
#Ready to recombine to pdf1 data
```

```python
#Renaming the columns to distiguish from the placeholder: 'Unnamed: '
newdf = pandas.DataFrame(remove_NA,columns =['State','2+ A1cTests'])
newdf
```

```python
#Dropping the column with state name and float value
df2 = newdf.join(pdf2_ind)
df2 = df2.drop('place',axis=1)
df2
#part2.drop('place',axis=1)
```

```python
#Renaming the columns of the second page of our table
df2=df2.rename(columns={'State':'State','2+ A1cTests':'2+ A1cTests','Unnamed: 0':'Foot Exm','Unnamed: 1':'Self-manage class','Unnamed: 4':'Dilate Eye','Unnamed: 7':'Bld glucose monitor'})
```

```python
#Combining into a single df
finaldf = pandas.concat([df1,df2],ignore_index=True)
finaldf
```

```python
#Saving to a csv file
finaldf.to_csv('Cleaned_data.csv',index=False)
```
